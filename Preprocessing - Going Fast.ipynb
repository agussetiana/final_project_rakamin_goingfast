# Data Preparation

## Outliers

```python
for num in nums:
  if num not in ['Administrative', 'BounceRates', 'ExitRates', 'SpecialDay','OperatingSystems', 'Browser', 'Region', 'TrafficType']:
    data_product2[num] = np.log(data_product2[num] + (data_product2[data_product2[num] > 0][num].min() / 2)) # kita harus pastikan data tidak 0 untuk dilogtransform
```

```python
import math

plt.figure(figsize=(12, 8))
for i in range (0, len(nums)):
  plt.subplot(math.ceil(len(nums)/3), 3, i+1)
  sns.violinplot(data=data_product2, y=nums[i], x='Revenue', linewidth=2, orient='v')
```

```python
from scipy import stats

print("Before removing outlier: ", len(data_product2))

for num in nums:
  if num not in ['id', 'rating']:
    z_scores = np.abs(stats.zscore(data_product2[num]))
    data = data_product2[z_scores <= 3]

print("After removing outlier: ", len(data_product2))
```

## Insight Outliers

Dengan melakukan transformasi log, Anda mencoba untuk mengurangi skewness atau perbedaan yang besar antara nilai-nilai dalam fitur tersebut. Transformasi log digunakan khususnya ketika data memiliki distribusi yang condong (skewed), di mana nilai-nilai terkonsentrasi pada salah satu ujung rentang nilai.

Dengan melakukan log transformation pada fitur-fitur yang memiliki outliers, Anda dapat mencoba untuk:

1. Mengurangi Pengaruh Outliers:

Outliers dapat memiliki pengaruh yang signifikan pada analisis statistik atau model pembelajaran mesin. Melakukan transformasi log dapat membantu mengurangi dampak outliers karena nilai-nilai ekstrem cenderung "ditarik" lebih dekat ke nilai-nilai tengah setelah transformasi.

2. Mengurangi Skewness:

Transformasi log sering digunakan untuk mengurangi skewness dalam distribusi data. Ketika data memiliki skewness yang signifikan, ini dapat menyebabkan kesalahan dalam analisis statistik dan pemodelan.

### Menambah Fitur Target berdasarkan Revenue

```python
data_product2['Target'] = data_product2['Revenue'].apply(lambda x: 'Buyer' if x == True else 'Non-Buyer')
```

```python
data_product2['Target'].unique()
```

```python
df_target_grp = data_product2.groupby(['Target']).agg(vist_cnt = ('Target', 'count')).sort_values('vist_cnt', ascending=True)
df_target_grp.reset_index(inplace=True)

df_target_grp['Percentage'] = round(df_target_grp['vist_cnt']/df_target_grp['vist_cnt'].sum()*100,2)
df_target_grp
```

```python
# Create barplot for Target
ax_diff = sns.barplot(x='Target', y='Percentage', data=df_target_grp, palette='OrRd')

# Annotate the bar with the value
for p in ax_diff.patches:
    ax_diff.annotate(f"{p.get_height():.2f}%", (p.get_x() + p.get_width() / 2., p.get_height()),
                     ha='center', va='center', fontsize=10, color='black', xytext=(0, 5),
                     textcoords='offset points')

# Formatting plot
plt.ylim(0, 100)
plt.title('Generate revenue by visitor status')
plt.xlabel('Visitor status',fontsize=12)
plt.ylabel('Percentage by Total Clients',fontsize=12)
plt.show()
```

## Handling Outliers

```python
df_outlier_train = pd.DataFrame(columns=['Features','Train_upper_outlier'])
df_outlier_test = pd.DataFrame(columns=['Features','Test_upper_outlier'])
```

```python
def outlier_values(df, df_output):
    for col in X_train:
        Q1 = df[col].quantile(0.25)
        Q3 = df[col].quantile(0.75)
        IQR = Q3 - Q1
        outlier = (Q3 + 1.5 * IQR)
        if outlier > 0:
            df_output.loc[len(df_output)] = [col, outlier.round(2)]
```

```python
outlier_values(X_train, df_outlier_train)
df_outlier_train
```

```python
outlier_values(X_test, df_outlier_test)
df_outlier_test
```

```python
# merge the two dataset
merged_outlier_df = pd.merge(df_outlier_train, df_outlier_test, on='Features', how='inner').sort_values(by=['Train_upper_outlier'], ascending=False)

merged_outlier_df
```

```python
df_perc_temp = pd.DataFrame(columns=['Features','Train_outlier_percentages','Test_outlier_percentages'])
```

```python
for index, row in merged_outlier_df.iterrows():
    train_outlier_val = row['Train_upper_outlier']
    test_outlier_val = row['Test_upper_outlier']
    
    train_outlier_count = X_train[row['Features']] > train_outlier_val
    test_outlier_count = X_test[row['Features']] > test_outlier_val
    
    train_outlier_percentage = round(train_outlier_count.sum()/len(X_train)*100,2)
    test_outlier_percentage = round(test_outlier_count.sum()/len(X_test)*100,2)
    
    df_perc_temp.loc[len(df_perc_temp)] = [row['Features'], train_outlier_percentage, test_outlier_percentage]
```

This indicates that a significant number of data points fall outside the expected range of the majority of the data in a column. This would suggests that the central tendency (mean, median) of the features is highly distorted.
